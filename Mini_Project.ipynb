```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Google

df1 = pd.read_csv('C:/Users/KIIT/Downloads/GOOG.csv')
```

```python
df1
```

```python
##2.


print(df1.head())


print(df1.isnull().sum())

df1 = df1.dropna()

df1['Date'] = pd.to_datetime(df1['Date'])

```

```python
##3.
df1['Daily_Return'] = df1['Close'].pct_change() * 100  


print(df1[['Close', 'Daily_Return']])
```

```python
df1
```

```python
###4.
plt.figure(figsize=(12, 6))
plt.plot(df1['Date'], df1['Volume'], label='Volume')
plt.title('Stock Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.show()
```

```python
##5.
window_size = 20

df1['SMA_20'] = df1['Close'].rolling(window=window_size).mean()


plt.figure(figsize=(12, 6))
plt.plot(df1['Date'], df1['Close'], label='Close Price')
plt.plot(df1['Date'], df1['SMA_20'], label=f'{window_size}-day SMA', color='orange')
plt.title('Stock Close Price with Moving Average')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
##6.
df1['Daily_Return'] = df1['Close'].pct_change() * 100
average_daily_return = df1['Daily_Return'].mean()
print(f"Average Daily Return: {average_daily_return:.2f}%")
```

```python
##7

df1['Trend'] = np.where(df1['Daily_Return'] > 0, 'Up', np.where(df1['Daily_Return'] < 0, 'Down', 'No Change'))
```

```python
df1
```

```python
##8.
import matplotlib.pyplot as plt


trend_counts = df1['Trend'].value_counts()


plt.figure(figsize=(8, 8))
plt.pie(trend_counts, labels=trend_counts.index, autopct='%1.1f%%', startangle=90, colors=['green', 'red', 'gray'])
plt.title('Trend Frequency')
plt.show()
```

```python
##correlation between the stocks AAPL,MSFT,GOOGL,AMZN
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df1['Ticker'] = 'AAPL'
df2['Ticker'] = 'MSFT'
df3['Ticker'] = 'GOOGL'
df4['Ticker'] = 'AMZN'


merged_df = pd.concat([df1, df2, df3, df4], ignore_index=True)


pivot_df = merged_df.pivot(index='Date', columns='Ticker', values='Daily_Return')

correlation_matrix = pivot_df.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix of Daily Returns')
plt.show()
```
